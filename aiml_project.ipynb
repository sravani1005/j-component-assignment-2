```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
d=pd.read_csv('/content/drive/MyDrive/aiml/archive (42).zip')
print(d)
```

```python
import seaborn as sns
sns.boxplot(data=d,x='ApplicantIncome' )
```

```python
from matplotlib import pyplot as plt
fig, ax=plt.subplots(1,2)
sns.histplot(data=d ,x=d['LoanAmount'], kde=True ,ax=ax[0] )
sns.boxplot(x=d['Credit_History'],y=d['Loan_Amount_Term'] ,ax=ax[1])
```

```python
corr_matrix=d.corr()
print(corr_matrix)
```

```python
cov_matrix=pd.DataFrame.cov(d)
sns.heatmap(cov_matrix,annot=True,fmt='g')
plt.show
```
